```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("churn.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
X = data.drop('Churn', axis = 1)
y = data['Churn']
```

```python
dropCols = ['customerID']
X.drop(dropCols, axis = 1, inplace = True)
```

```python
catCol = []
for i in X.columns:
    if X[i].dtype == 'object':
        catCol.append(i)
catCol
```

```python
X['TotalCharges'].value_counts()
```

```python
pd.f
```

```python
# s = 0
# n = 0
totalCharges = []
idx = 0
errors = []
for i in X['TotalCharges']:
    try:
        totalCharges.append(float(i))
#         s = s + float(i)
#         n += 1
    except:
        errors.append(idx)
    idx = idx + 1
```

```python
errors
```

```python
XNew = X.drop(errors, axis = 0)
```

```python
totalCharges
```

```python
XNew.shape
```

```python
len(totalCharges)
```

```python
XNew['TotalCharges'] = totalCharges
```

```python
XNew['TotalCharges']
```

```python
y = y.replace({'Yes': 1, 'No' : 0})
```

```python
y.shape
```

```python
yNew = y.drop(errors, axis = 0)
```

```python
yNew.shape
```

```python
XNew.head()
```

```python
XFinal = pd.get_dummies(XNew, drop_first=True)
```

```python
#Normalization
#val - val.mean() / val.std()
```

```python
XFinal.describe().T
```

```python
XFinal = (XFinal - XFinal.mean()) / XFinal.std() #Normalization
```

```python
XFinal['labels'] = yNew
```

```python
corrMat = XFinal.corr()
```

```python
corrMat = corrMat['labels'].sort_values(ascending = False)
```

```python
corrMat.items
```

```python
dropCols = ['labels']
for i, v in corrMat.items():
    if abs(v) < 0.1:
        dropCols.append(i)
dropCols
```

```python
XFinal.drop(dropCols, axis = 1, inplace = True)
```

```python
X_train, X_val, y_train, y_val = train_test_split(XFinal, yNew, test_size = 0.2, random_state = 1)
```

```python
X_train.head()
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
#Actual testing
pred = lr.predict(X_val)
```

```python
print(classification_report(pred, y_val))
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
pred = svc.predict(X_val)
```

```python
print(classification_report(pred, y_val))
```

```python
X_val[pred == 1]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
pred = rfc.predict(X_val)
```

```python
print(classification_report(pred, y_val))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
pred = dt.predict(X_val)
```

```python
print(classification_report(pred, y_val))
```
